## Dataset

```python
# DO NOT MODIFY
from sklearn.datasets import make_moons

X, Y = make_moons(random_state=42, n_samples=(50, 450), noise=0.25)
```

## Training

```python
# create train/test variables
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=.3, random_state=42)
```

```python
# use gaussian naive bayes
from sklearn.ensemble import RandomForestClassifier as rfc
model = rfc(n_estimators=100, random_state=42)
model.fit(X_train, Y_train)
```

## Testing / Optimization

```python
# test model on X_test
Y_pred = model.predict(X_test)
```

```python
# use accuracy score (# correct predictions / total # predictions) as performance metric
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(Y_test, Y_pred)
accuracy
```

```python
# evaluate model on X_train to check for overfitting
Y_train_pred = model.predict(X_train)
```

```python
train_accuracy = accuracy_score(Y_train, Y_train_pred)
train_accuracy
```
